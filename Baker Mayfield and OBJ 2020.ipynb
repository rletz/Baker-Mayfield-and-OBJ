# Did Baker Mayfield Actually Get Better Without Odell Beckham Jr.?

October 25th, 2020. The Cleveland Browns are in Cincinnati, playing the last place, division rival Bengals at Paul Brown Stadium. The Browns, sporting a 4-2 record to this point in the season, are feeling confident about the way they are playing, despite their recent 38-7 blowout loss to rival Pittsburgh at Heinz Field the previous week.

Midway through the first quarter, Baker Mayfield throws his only interception of the game and Odell Beckham Jr., Cleveland's star wide receiver, makes an attempt to tackle the defending ball carrier.

This would be the last time the Browns have OBJ on the field for the remainder of the 2020 season.

Beckham Jr. would not return to the game with a knee injury. Further examination the following day would cement initial fears that Beckham Jr. had torn the ACL in his left knee. The Browns placed him on IR for the rest of the year, ending his season.

There were concerns that Mayfield and the Browns may not be able to continue to perform to their season standard without their star receiver. There's no doubting Beckham Jr.'s talent, as he has a track record to prove he is one of the better receivers in the NFL. If people were going to doubt Baker Mayfield and the Browns, those doubts would not be unreasonable.

However, as we know now, those concerns following their Week 7 win proved to be for nothing. The Browns finished with an 11-5 record, finishing 3rd in the AFC North and, more importantly to fans in Cleveland, clinched a postseason spot for the first time in 18 years. 

Now that the season is over, I was curious to go back through Baker Mayfield's season and get answers to these questions:

Was Baker Mayfield impacted by losing Odell Beckham Jr. during the season? 

And could it be possible that Baker actually ***performed better*** without OBJ in the lineup?

NOTE: Please note that the data used below was pulled from www.pro-football-reference.com.

```python
##importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read file
baker = pd.read_csv('C:/Users/Ryan/Desktop/Python/Baker Mayfield/Baker Mayfield 2020.csv')
```

```python
#setting variables
before_injury = baker[baker['Week'] < 8]
after_injury = baker[baker['Week'] >= 8]
before_injury_mean = before_injury['Rate'].mean()
after_injury_mean = after_injury['Rate'].mean()
before_injury_ints = before_injury['Int'].sum()
after_injury_ints = after_injury['Int'].sum()
before_injury_atts = before_injury['Att'].mean()
after_injury_atts = after_injury['Att'].mean()
```

## Analyzing Baker Mayfield's Passer Rating With and Without Odell Beckham Jr.

We are going to use Baker Mayfield's passer rating to determine his performance throughout the season, which is compiled by taking a quarterback's pass attempts, pass completions, touchdowns and interceptions thrown, and passing yardage totals.

We will also analyze Mayfield's interception totals, as well as his pass attempt averages, both before and after the injury for possible trends.

### Weekly

```python
#creating line chart of Baker's passer ratings per week
plt.figure(figsize=(15,9))

plt.plot(baker['Week'], baker['Rate'],c='saddlebrown', label='Passer Rating', linewidth=2.5)
#set labels and limits
plt.xlabel('Week')
plt.ylabel('Passer Rating')
plt.xlim(1,17)
plt.ylim(50,158.3)
#add vertical line for OBJ week injured
plt.axvline(7, c='red', label = 'Beckham Jr. Injury', alpha=0.5)
#add vertical line for bye week (week 9)
plt.axvline(9, c='black', label='Bye Week', alpha=0.5)
#add horizontal line for season average pass rating
plt.axhline(baker['Rate'].mean(), c='blue', label='Mayfield Average QBR 2020', alpha=0.1)
#set title and legend
plt.title('Baker Mayfield 2020 Season - Passer Rating')
plt.legend(loc='upper left')
plt.show()
```

### Average QBR Before and After OBJ Injury

```python
print('Baker Mayfield - QBR (Quarterback Rating) Averages')
print('\n')
print('QBR Before Odell Beckham Jr. Injury : ', round(before_injury_mean,2))
print('----------------------------------------------')
print('QBR After Odell Beckham Jr. Injury : ', round(after_injury_mean,2))
```

## Analyzing Baker Mayfield's Interceptions Per Game

```python
#creating line chart of Mayfield's interception output per game
fig = plt.figure(figsize=(14,7))

plt.plot(baker['Week'], baker['Int'], color='coral', label='Interceptions', linewidth=3)
#configuring x and y axes
plt.xlabel('Week')
plt.ylabel('Interceptions by Game')
plt.xlim(1,17)
plt.ylim(0,3)
plt.yticks(range(0,3))
#setting OBJ injury line
plt.axvline(7, c='red', label='Beckham Jr. Injury', alpha=0.5)
#setting bye week line
plt.axvline(9, c='black', label='Bye Week', alpha=0.5)
plt.title('Baker Mayfield 2020 Season - Interceptions / Game')
plt.legend(loc='upper right')
plt.show()
```

## Baker Mayfield's Passing Attempts Per Game

```python
print('Baker Mayfield - Pass Attempt Averages')
print('\n')
print('Average Pass Attempts Before Beckham Jr. Injury : ', round(before_injury_atts, 0))
print('-------------------------------------------------------')
print('Average Pass Attempts After Beckham Jr. Injury : ', round(after_injury_atts, 0))
```

## Observations

### Heating Up

As evidenced in the **"Average QBR Before and After Injury"** section, we can see that Baker Mayfield's passer rating actually ***increased*** once Odell Beckham Jr. was ruled out for the season following his knee injury. Mayfield's rating may not have increased dramatically, but it did still increase a little over 3 points without OBJ.

This may serve as a surprise judging that Mayfield didn't just lose any wide receiver, but his ***star receiver,*** and his QBR still managed to increase without him in the lineup.

Also, after looking at the graph above in the **"Weekly"** section, we notice that 2 of Mayfield's best games in terms of QBR came after Beckham Jr.'s injury.

His best QBR of the season came in Week 13, long after Beckham Jr. was sidelined for the season. He followed that up two weeks later with his third best performance in Week 15.

Interestingly, Baker's second best QBR of the season came in Week 7, the same week in which Odell was injured. What makes that particular passer rating stand out?

Beckham Jr. was only targeted once during the entire game. It was on the intercepted pass midway through the first quarter that Odell was later injured on.

### Ball Security

After reviewing the graph, **"Baker Mayfield 2020 Season - Interceptions/Game"**, we learned that something did decrease from Baker Mayfield's game, just maybe not what we expected. It was not Mayfield's QBR, but rather the ***number of interceptions*** he threw, that dropped as a result.

Before Odell's injury, Baker may have been forcing the ball a little too much in his direction, causing dangerous throws or, possibly even, interceptions.

Mayfield only threw ***one interception*** in the 9 games following OBJ's injury, a sharp contrast from the 7 interceptions he threw with him suiting up.

### Keep the Arm Loose

In relation to noticing that the interceptions thrown by Mayfield dropped dramatically without Odell in the lineup, we can also note from the section **"Baker Mayfield's Passing Attempts Per Game"** that Mayfield averaged about 4 more passes a game without him as well.

While it is certainly impressive that the Heisman Trophy winner from Oklahoma only threw one interception in the 9 games without Beckham Jr., it makes it even more noteworthy because he was able to limit his interceptions while throwing the ball more often.

## Conclusion

In conclusion, after analyzing all of the above data, it does show that Baker Mayfield actually did improve without having Odell Beckham Jr. in the lineup for the rest of the 2020 season.

Not only did his QBR spike about 3 points without him, he also dramatically limited his interceptions while throwing the ball more.

Does this mean that the Cleveland Browns would be a better football team without Odell Beckham Jr. on the roster? Hard to tell. Again, there is no doubting the talent that OBJ has and the explosiveness he can provide to an offense when he steps between the lines.

However, something perhaps even more remarkable than Mayfield's play without his star receiver is worth mentioning. Baker was able to lead the Cleveland Browns to their first postseason win since 1994, defeating their hated rival Pittsburgh Steelers in Pittsburgh in the Wild Card round, 49-38.

Odell Beckham Jr. was not on the field for one snap of the victory.
